Calcular la desviación de Unidades vendidas

Usar value_counts para saber la categoría más frecuente

PrecioReal crear columna, cuál es el producto más caro
UnitPrice_COP / (1+ InflacionRate_pct /100) -> Usar loc[] para obtener el nombre del producto

```python
desviación=df["UnitsSold"].std()
print(f"Desviación estándar de las unidades vendidas: {desviación}")
cat_freq = df["Category"].value_counts().idxmax()
cont_freq = df["Category"].value_counts().max()

print(f"La categoría más frecuente es: {cat_freq} con {cont_freq} ventas registradas")
print(df.columns)
df["PrecioReal"] = df["UnitPrice_COP"]/(1+ df["InflationRate_pct"]/100)
pr_pr_alt = df.loc[df["PrecioReal"].idxmax(), "ProductID"]
pr_alt= df["PrecioReal"].max()

print(f"El producto con el precio real más alto es {pr_pr_alt} con un valor de {pr_alt:.2f}")
```
