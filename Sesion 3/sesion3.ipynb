```python
def salario_neto(salario_b):
  impuesto=0

  if salario_b <= 40000:
    impuesto=0
  elif salario_b <=80000:
    impuesto=(salario_b-40000)*0.1
  elif salario_b <=120000:
    impuesto = (4000+(salario_b-80000)*0.2)
  else:
    impuesto=(4000+8000+(salario_b-120000)*0.3)

  salario_neto=salario_b-impuesto

  return salario_neto

salario=int(input("Introduzca su salario"))
print(f"Su salario neto es: {salario_neto(salario)}")
```
