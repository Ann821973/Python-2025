#Fundamentos de Python

```python
q0,q1 = 100,80
p0,p1= 50,55

per_dQ=(q1-q0)/q0
per_dP=(p1-p0)/p0
el=per_dQ/per_dP

clasif= "elastica" if abs(el)>1 else "inelastica" if abs(el)<1 else "unitaria"
print(f"Elastidad: {el:.3f} ({clasif})")
```

```python
print("ingrese una cantidad 1")
cantidad1=int(input())
print("ingrese una cantidad 2")
cantidad2=int(input())

precio1=int(input("ingrese un precio 1"))
precio2=int(input("ingrese un precio 2"))

per_dq=(cantidad2-cantidad1)/cantidad1
per_dp=(precio2-precio1)/precio1
el=per_dq/per_dp

clasif= "elastica" if abs(el)>1 else "inelastica" if abs(el)<1 else "unitaria"
print(f"Elastidad: {el:.3f} ({clasif})")

```
